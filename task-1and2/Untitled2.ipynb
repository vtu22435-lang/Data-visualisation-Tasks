```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
sns.set(style="whitegrid")
uploaded = files.upload()
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)
print(" First 5 rows of the dataset:")
print(df.head())
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
# df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce') # This column does not exist
print("\n Missing Values in Each Column:")
print(df.isnull().sum())
print("\n Data Types of Columns:")
print(df.dtypes)
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Product line', order=df['Product line'].value_counts().index)
plt.title("Bar Chart - Frequency of Product Lines")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
region_counts = df['Branch'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%',
startangle=140)
plt.title("Pie Chart - Branch Distribution")
plt.axis('equal')
plt.show()
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Sales', y='gross income', hue='Branch')
plt.title("Scatterplot - Sales vs Gross Income")
plt.show()
df_sorted = df.sort_values('Date')
plt.figure(figsize=(12,6))
sns.lineplot(data=df_sorted, x='Date', y='Sales')
plt.title("Line Plot - Sales Over Time")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.figure(figsize=(10,6))
sns.stripplot(data=df, x='Branch', y='gross income', jitter=True)
plt.title("Strip Plot - Gross Income by Branch")
plt.xticks(rotation=45)
plt.show()
plt.figure(figsize=(12,6))
sns.swarmplot(data=df, x='Product line', y='Sales')
plt.title("Swarm Plot - Sales by Product Line")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.figure(figsize=(8,6))
sns.histplot(df['Sales'], bins=20)
plt.title("Histogram - Sales")
plt.show()
plt.figure(figsize=(8,6))
sns.kdeplot(df['gross income'], fill=True)
plt.title("Density Plot - Gross Income")
plt.show()
plt.figure(figsize=(8,6))
sns.rugplot(df['Sales'])
plt.title("Rug Plot - Sales")
plt.show()
```
